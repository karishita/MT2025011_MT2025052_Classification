```python
iqr_bounds = {}
for col in num_cols:
    Q1 = X_train[col].quantile(0.25)
    Q3 = X_train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    iqr_bounds[col] = (lower, upper)
    print(f"{col}: Q1={Q1:.3f}, Q3={Q3:.3f}, IQR={IQR:.3f}, lower={lower:.3f}, upper={upper:.3f}")
```

```python
#Flag outliers (rows that are outside bounds for any numeric column)
outlier_mask = np.zeros(len(X_train), dtype=bool)  # False => keep
for col in num_cols:
    lower, upper = iqr_bounds[col]
    mask_col = (X_train[col] < lower) | (X_train[col] > upper)
    print(f"Outliers in {col}: {mask_col.sum()} rows")
    outlier_mask = outlier_mask | mask_col

print("Total rows flagged as outlier (train):", outlier_mask.sum())
print("Proportion removed:", outlier_mask.sum() / len(X_train))
```
